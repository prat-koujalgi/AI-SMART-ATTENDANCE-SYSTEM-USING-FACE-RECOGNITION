Here I am creating a Pandas DataFrame named 'df' with 4 features and 4 observations, using random values generated by NumPy:

Task 1:

```python
import numpy as np
import pandas as pd

np.random.seed(42)

data = np.random.rand(4, 4)

df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3', 'Feature4'])
```

We can access and manipulate the data in the DataFrame using various Pandas operations.

Task 2: Rename column names

```python
df.columns = ['Random value 1', 'Random value 2', 'Random value 3', 'Random value 4']
```

Task 3: Descriptive statistics

```python
statistics = df.describe()
print(statistics)
```

Task 4: Check for null values and data types

```python
null_counts = df.isnull().sum()
data_types = df.dtypes

print("Null Value Counts:")
print(null_counts)

print("\nData Types:")
print(data_types)
```

Task 5: Display columns using location method and index location method

```python
print(df.iloc[:, 1:3])
print(df[['Random value 2', 'Random value 3']])
```
